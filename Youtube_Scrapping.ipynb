```
from bs4 import BeautifulSoup #for scraping
import requests               #required for reading the file
import pandas as pd           #(optional) Pandas for dataframes 
import json                   #(optional) If you want to export json
import os
import re
```

```
def retrieve_info(url,meta):
  page = requests.get(url)
  soup = BeautifulSoup(page.text, 'html.parser')
  #title
  title_element=soup.find("span",{"class" : "watch-title"}).get_text()
  #category
  category_element=soup.find("a",{"class" : " yt-uix-sessionlink spf-link "}).get_text()
  #description
  description_element=soup.find("meta",attrs={'name':"description"}).get('content')
  #No_subscribers
  subs_element=soup.find("span",{"class" : "yt-subscription-button-subscriber-count-branded-horizontal yt-subscriber-count"}).get_text()
  #uploaded_by
  uploader_element=soup.find("a",{"class" : "yt-uix-sessionlink spf-link ",}).get_text()
  #date_uploaded
  date_element=soup.find("meta",attrs={'itemprop':"datePublished"}).get('content')
  #dislikes
  dislikes_element=soup.find("button",{"title" : "I dislike this"}).get_text()
  #likes
  likes_element=soup.find("button",{"title" : "I like this"}).get_text()
  #views
  views_element=soup.find("div",{"class" : "watch-view-count"}).get_text()
  #tags
  tag_element=soup.find("span",{"class" : "standalone-collection-badge-renderer-text"}).get_text()

  meta['title']=title_element
  meta['category']=category_element
  meta['description']=description_element
  meta['subscribers']=subs_element
  meta['user']=uploader_element
  meta['date']=date_element
  meta['dislikes']=dislikes_element
  meta['likes']=likes_element
  meta['views']=views_element
  meta['tags']=tag_element

  return meta
```

```
def get_related_links(url):
  source = requests.get(url).text
  soup = BeautifulSoup(source, 'html.parser')
  souppy=soup.findAll("li", {"class":"video-list-item related-list-item show-video-time related-list-item-compact-video"})

  final_list=[]

  for ih in range(len(souppy)):
    souppy1=BeautifulSoup(str(souppy[ih]),'html.parser')
    url=souppy1.find("a",{"class":" content-link spf-link yt-uix-sessionlink spf-link "}).get('href')
    duration=souppy1.find("span",{"class":"accessible-description"}).get_text()
    thumbnail=souppy1.find("img").get('data-thumb')

    meta={}
    meta['url']=str("https://www.youtube.com/"+url)
    meta['duration']=duration
    meta['thumbnail']=thumbnail

    final_list.append(meta)
  
  return final_list
```

```
def populate(url):
  info_list=[]
  i=0
  while len(info_list)<1000:
    list1=get_related_links(url)
    info_list=info_list+list1
    url=info_list[i]['url']
    print(i)
    i+=1
  
  return info_list
```
