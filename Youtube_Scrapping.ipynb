```
from bs4 import BeautifulSoup #for scraping
import requests               #required for reading the file
import pandas as pd           #(optional) Pandas for dataframes 
import json                   #(optional) If you want to export json
import os
import re
```

```
def get_related_links(url):
  source = requests.get(url).text
  soup = BeautifulSoup(source, 'html.parser')
  souppy=soup.findAll("li", {"class":"video-list-item related-list-item show-video-time related-list-item-compact-video"})

  final_list=[]

  for ih in range(len(souppy)):
    souppy1=BeautifulSoup(str(souppy[ih]),'html.parser')
    url=souppy1.find("a",{"class":" content-link spf-link yt-uix-sessionlink spf-link "}).get('href')
    duration=souppy1.find("span",{"class":"accessible-description"}).get_text()
    thumbnail=souppy1.find("img").get('data-thumb')

    meta={}
    meta['url']=str("https://www.youtube.com/"+url)
    meta['duration']=duration
    meta['thumbnail']=thumbnail

    final_list.append(meta)
  
  return final_list
```

```
def populate(url):
  info_list=[]
  i=0
  while len(info_list)<1000:
    list1=get_related_links(url)
    info_list=info_list+list1

    if(len(info_list)!=0):
      url=info_list[i]['url']
      i+=1
    else:
      break;
      
    print("i=",i," info_list=",len(info_list)," list1=",len(list1))
    
  return info_list
```

```
def retrieve_info(url,meta):
  page = requests.get(url)
  try:
    if(page.status_code)!=200:
      return meta
    soup = BeautifulSoup(page.text, 'html.parser')

    #title
    title_element=soup.find("span",{"class" : "watch-title"}).get('title')
    #category
    category_element=soup.find("a",{"class" : " yt-uix-sessionlink spf-link "})
    if category_element is None:
      category_element="Null"
    else:
      category_element=category_element.get_text()
    #description
    description_element=soup.find("meta",attrs={'name':"description"})
    if description_element is None:
      description_element="Null"
    else:
      description_element=description_element.get('content')

    #No_subscribers
    subs_element=soup.find("span",{"class" : "yt-subscription-button-subscriber-count-branded-horizontal yt-subscriber-count"})
    if subs_element is None:
      subs_element=0
    else:
      subs_element=subs_element.get('title')

    #uploaded_by
    uploader_element=soup.find("a",{"class" : "yt-uix-sessionlink spf-link ",}).get_text()
    #date_uploaded
    date_element=soup.find("meta",attrs={'itemprop':"datePublished"}).get('content')
    #dislikes
    dislikes_element=soup.find("button",{"title" : "I dislike this"})
    if dislikes_element is None:
      dislikes_element=0
    else:
      dislikes_element=dislikes_element.get_text()
    
    #likes
    likes_element=soup.find("button",{"title" : "I like this"})
    if likes_element is None:
      likes_element=0
    else:
      likes_element=likes_element.get_text()
    
    #views
    views_element=soup.find("div",{"class" : "watch-view-count"})
    if views_element is None:
      views_element=0
    else:
      views_element=views_element.get_text()
    #tags
    tag_element=soup.find("span",{"class" : "standalone-collection-badge-renderer-text"})
    if tag_element is None:
      tag_element="Null"
    else:
      tag_element=tag_element.get_text()

    meta['title']=title_element
    meta['category']=category_element
    meta['description']=description_element
    meta['subscribers']=subs_element
    meta['user']=uploader_element
    meta['date']=date_element
    meta['dislikes']=dislikes_element
    meta['likes']=likes_element
    meta['views']=views_element
    meta['tags']=tag_element

    return meta

  except:
    return retrieve_info(url,meta)
```

```
import csv
keys = ['title','category','description','subscribers','user','date','dislikes','likes','views','tags','url','thumbnail','duration']
output_file=open('youtube_dataset_01.csv', 'w')
dict_writer = csv.DictWriter(output_file,fieldnames=keys)
dict_writer.writeheader()
```

```
list_info_02=[]
index=0

# loop through each of the dict items of list_universal
for item in list_universal:
  url=item['url']
  index+=1
  print(index," ",url)
  # retrieve info to get the final dict
  final_dict=retrieve_info(url,item)
  list_info_02.append(final_dict)

  if(index%100==0):
    print("Writing to the File")
    dict_writer.writerows(list_info_02)
    list_info_02=[]
```

```
if len(list_info_02)!=0:
  print("Writing to the File")
  dict_writer.writerows(list_info_02)
  
output_file.close()
```
